**Mục tiêu**: Tích hợp 6 module (M1-M6) từ các bài trước (1-11) thành một hệ thống AIDEOM-VN đầy đủ

## I. Kiến Trúc Hệ Thống AIDEOM-VN

### 1. Sáu Module Thành Phần

| Module | Tên | Kỹ Thuật | Input | Output |
|--------|-----|---------|-------|--------|
| **M1** | Dự báo kinh tế | Cobb-Douglas + VAR | GDP, TFP, lao động 2020-2025 | Dự báo 2026-2030 |
| **M2** | Đánh giá sẵn sàng | TOPSIS + Entropy | Chỉ số 6 vùng, 10 ngành | Xếp hạng theo AI Readiness |
| **M3** | Tối ưu phân bổ | LP/MIP + Dynamic | Ngân sách, hiệu quả, ràng buộc | Phân bổ tối ưu vùng-thời gian |
| **M4** | Mô phỏng lao động | Markov + Automation | Kế hoạch AI/H, risk | NetJob từng ngành |
| **M5** | Đánh giá rủi ro | Stochastic LP + Scenario | Tham số rủi ro, kịch bản | Cyber, Env, Regional gini |
| **M6** | Dashboard | Streamlit + Plotly | M1-M5 outputs | Trực quan 5 kịch bản |

### 2. Luồng Dữ Liệu

```
CSV Data (vietnam_macro, regions, sectors)
    ↓
M1: Forecast GDP 2026-2030
M2: Assess Regional Readiness
    ↓
    ├→ M3: Allocate Budget (5 scenarios)
    ├→ M4: Simulate Labor Market
    └→ M5: Assess Risks
    ↓
M6: Streamlit Dashboard
    ↓
KPI Comparison | Allocation Heatmap | Risk Alert
```

### 3. Năm Kịch Bản Chính Sách

1. **S1 - Truyền thống**: 70% K + 10% D/AI/H (GDP cao, rủi ro công nghệ cao)
2. **S2 - Số hóa nhanh**: 25% K + 45% D + 15% AI + 15% H (Tăng trưởng bền vững)
3. **S3 - AI dẫn dắt**: 20% K + 20% D + 45% AI + 15% H (GDP cao nhất, phụ thuộc AI)
4. **S4 - Bao trùm số**: 30% K + 20% D + 10% AI + 40% H (Giảm bất bình đẳng, nhiều việc làm)
5. **S5 - Tối ưu cân bằng**: 35% K + 25% D + 20% AI + 20% H (Cân bằng 4 mục tiêu)
